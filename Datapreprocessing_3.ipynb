# REPLACE WITH MEAN VALUES OF NAN VALUES

```python
ds[ds.loc[:,'LotConfig']=='Inside']['LotFrontage'].replace(np.nan,ds[ds.loc[:,'LotConfig']=='Inside']['LotFrontage'].mean())
```

# now fill all missing values with respect to class like(inside , FR2,FR3,cornor etc)

```python
ds_copy = ds.copy()
for var_class in ds['LotConfig'].unique():
    ds_copy.update(ds[ds.loc[:,'LotConfig']==var_class]['LotFrontage'].replace(np.nan,ds[ds.loc[:,'LotConfig']==var_class]['LotFrontage'].mean()))
```

ds_copy.isnull().sum()

FILL THE MISSING VALUES WITH DOMAIN VALUES

```python
ds_copy = ds.copy()
num_vars_miss = ['LotFrontage','MasVnrArea','GarageYrBlt']
cat_vars = ['LotConfig','MasVnrType','GarageType']
for cat_var , num_var_miss in zip(cat_vars , num_vars_miss):
    for var_class in ds[cat_var].unique():
        ds_copy.update(ds[ds.loc[:,cat_var]==var_class][num_var_miss].replace(np.nan,ds[ds.loc[:,cat_var]==var_class][num_var_miss].mean()))
```

```python
ds_copy[num_vars_miss].isnull().sum()
```

```python
ds_copy[ds_copy[['MasVnrType']].isnull().any(axis =1)]
```

```python
ds_copy[ds_copy[['GarageType']].isnull().any(axis =1)]
```

fill the NAN values with MEAN

```python
ds_copy = ds.copy()
num_vars_miss = ['LotFrontage','MasVnrArea','GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var , num_var_miss in zip(cat_vars , num_vars_miss):
    for var_class in ds[cat_var].unique():
        ds_copy.update(ds[ds.loc[:,cat_var]==var_class][num_var_miss].replace(np.nan,ds[ds.loc[:,cat_var]==var_class][num_var_miss].mean()))
```

```python
ds_copy[num_vars_miss].isnull().sum()
```

Data Distribution

```python
plt.figure(figsize = (10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(ds[var],bins = 20, kde_kws = {'linewidth':5 , 'color':'r'},label = 'original datasets')
    sns.distplot(ds_copy[var],bins = 20, kde_kws = {'linewidth':5 , 'color':'b'},label = 'mean replaced datasets')
    plt.legend()
```

Now fill the nan values with median

```python
ds_copy_median = ds.copy()
num_vars_miss = ['LotFrontage','MasVnrArea','GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var , num_var_miss in zip(cat_vars , num_vars_miss):
    for var_class in ds[cat_var].unique():
        ds_copy_median.update(ds[ds.loc[:,cat_var]==var_class][num_var_miss].replace(np.nan,ds[ds.loc[:,cat_var]==var_class][num_var_miss].median()))
```

```python
ds_copy_median[num_vars_miss].isnull().sum()
```

Data Oistribution

```python
plt.figure(figsize = (10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(ds[var],bins = 20, kde_kws = {'linewidth':5 , 'color':'r'},label = 'original datasets')
    sns.distplot(ds_copy[var],bins = 20, kde_kws = {'linewidth':5 , 'color':'green'},label = 'mean value')
    sns.distplot(ds_copy_median[var],bins = 20, kde_kws = {'linewidth':5 , 'color':'b'},label = 'median values')
    plt.legend()
```

Boxplot for checking outlier

```python
for i, var in enumerate(num_vars_miss):
    plt.figure(figsize = (10,10))
    plt.subplot(3,1,1)
    sns.boxplot(ds[var])
    plt.subplot(3,1,2)
    sns.boxplot(ds_copy[var])
    plt.subplot(3,1,3)
    sns.boxplot(ds_copy_median[var])
```
