# checking only catagorical vaules

```python
cat_vars = ds.select_dtypes(include = 'object')
cat_vars.tail()
```

```python
miss_null_vars = cat_vars.isnull().mean()*100
miss_null_vars
```

```python
drop_vars=['Alley','FireplaceQu','PoolQC','Fence','MiscFeature']
cat_vars.drop(columns=drop_vars, axis=1,inplace = True)
cat_vars.shape
```

```python
isnull_per = cat_vars.isnull().mean()*100
miss_var = isnull_per[isnull_per>0].keys()
miss_var
```

```python
cat_vars['MasVnrType'].mode() #checking missing values
```

```python
cat_vars['MasVnrType'].value_counts() #count of none values
```

```python
cat_vars['MasVnrType'].fillna(cat_vars['MasVnrType'].mode()[0])
```

```python
for var in miss_var:
    cat_vars[var].fillna(cat_vars[var].mode()[0],inplace =True)
    print(var,'=',cat_vars[var].mode()[0])
    
```

```python
cat_vars.isnull().sum()
```

```python
plt.figure(figsize = (10,10))
for i, var in enumerate(miss_var):
    plt.subplot(4,3,i+1)
    plt.hist(cat_vars[var],label = 'Impute')
    plt.hist(ds[var].dropna(),label = 'original')
    plt.legend()
```

```python
ds.update(cat_vars)
ds.drop(columns =drop_vars,inplace =True)
```
